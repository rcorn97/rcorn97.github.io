```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dfconstruction = pd.read_csv('ConstructionTimeSeriesDatav2.csv')
Year = []
for i in range(len(dfconstruction['Month-Year'])):
    Year.append(2000+int(dfconstruction['Month-Year'][i][-2:]))
dfconstruction['Year'] = Year
dfconstruction
```

```python
#Graph
fig, ax = plt.subplots()
ax.plot(dfconstruction['Month-Year'],dfconstruction['Private Construction'], color = 'g')
ax.plot(dfconstruction['Month-Year'],dfconstruction['Public Construction'], color = 'y')

#Axis Enhancements
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xticks(dfconstruction['Month-Year'][::20])
ax.set_xticklabels(dfconstruction['Month-Year'][::20])
ax.set_xlabel('Month-Year',fontsize='12',fontname='Times New Roman')
ax.set_ylabel('Dollars in Millions ($)',fontsize='12',fontname='Times New Roman')
ax.set_ylim(0,max(dfconstruction['Private Construction'])+10000)
ax.axvspan(55,109,ymin=0,ymax = 11000,facecolor='k',alpha=0.04)
ax.axvline(x=55,linewidth=1,color='k',label='Largest Variance',linestyle='-')
ax.axvline(x=109,linewidth=2,color='k',label='Smallest Variance',linestyle='--')
ax.legend(loc=4,fontsize='8')

#Text Enhancements
ax.text(0,68000,'Private Construction', color='g', fontsize='8')
ax.text(0,27000,'Public Construction', color='y', fontsize='8')
ax.text(57.5,90000, '<------ Housing Market Crisis ------>', color='k', fontsize='10')

#Figure Enhancements
fig.set_size_inches(10,5)
fig.suptitle("Housing Market's Impact on Private vs. Public Construction Value",fontsize='14',fontname='Times New Roman')
```

The graph above shows the impact of the United States housing market crash, highlighted by the shaded area, and it's impact on Public vs. Private construction. Before the housing market crash, Private concstruction significantly out paced Public construction with the solid straight line on the left representing the largest variance between the two. However, as the hosuing market crisis started to unfold in about 2007, Public and Private construction formed an inverse relationship where as Private construction decreased, Public increased. The variance between the two reached its lowest in 2011, represented by the dotted line on the right. As the nation began to recover economically, Public construction began to recoup its dominance. Therefore, in times of economic hardships, such as the housing market crash above, the Private sector is more susceptible to performing worse than the Public sector.
